```python
#Examing two attributes for outliers   ---   6. Examine vacant.location and 
#Explanation:
#The reason I chose 'Districts' is because if the district is not in the list and isn't a NaN.  Then the location could be 
#out of baltimore.  Which if this was the case would not be valid in determining if vacancies affect crime. 
districts = ["CENTRAL", "SOUTHEASTERN", "EASTERN", "NORTHEASTERN", "NORTHERN", "NORTHWESTERN", "WESTERN", "SOUTHWESTERN", "SOUTHERN"]

#Outliers exist in Districts for crime
False in crime.District.isin(districts)
#append index to array for human inspection
       
```

```python
#Check if long-lat are within the range of baltimore
#Explanation:
#The reason I chose Location is for checking if the long-lat are within the range of baltimore.  If these values are different
#integers than (39, -76) then the data is out of baltimore and also should not be included. 
#
index = 0
outliers = []
for item in list(vacant.Location):
    loc = item.strip("()\n").split(',')
    if int(float(loc[0])) != 39 or int(float(loc[1])) != -76:
        print("Outlier @ index: ", index)
        outliers.append(index)
    index += 1
if len(outliers):
    print("Outliers list")
    outliers
else:
    print("No outliers")
```

```python
#Reducing Numerosity   ---   9. Reduce dates to ordinal integers then bin per month, Jan = 1, ... ,Dec = 12.
#The reason why I picked binning is because this could be a good way to allocate data per month. 
#An alternative I was looking at visually was for Histogramming the dates with 12 bins.  This way 
#the crimes would be spread out over 12 months
binned_dates = []
vacant_date = pd.DataFrame(columns=['NoticeDate'])
vacant_date["NoticeDate"] = pd.to_datetime(vacant["NoticeDate"])
for i in range(0, len(vacant_date), 1):
    binned_dates.append(vacant_date["NoticeDate"][i].month)
binned_dates[0:20]
```

```python
#Date Reduction   ---   11. Sampling, with, without replacement.  Stratified & unstratified

size = 100

#Without replacement
sample = crime.sample(size, replace=False)

#With replacement
sample_replace = crime.sample(size, replace=True)

#Stratified sampling
sample_strat = []
step = m.ceil(len(crime) * .005)
for i in range(0, 200, 1):
    sample_strat.append(crime[(i * step):((i + 1) * step)].sample(n=1))

#Unstratified sampling
sample_unstrat = crime.sample(size)
```

```python
#Contingency Table and Chi-square   ---   12. 

# Contingency table
contingency = pd.crosstab(vacant.Block, crime.Weapon)


# Chi^2 test of independence.
alpha = .05
#c, p, dof, expected
results = chi2_contingency(contingency)

#Print P value compared against Alpha
results[1] < alpha

#Explain Findings:
#My finds with doing the contigency table is I set up a null hypothesis.  Which states that the vacant block and weapon
#used are not correlated.  I set my alpha limit to .05.  The contigency is checked and stored in results.
#I check the P-value which is used to reject or confirm my hypothesis.  If the result is true then we can
#reject the null hypothesis and assert that there's a correlation between the block and weapon used.  
#in this case, since the P value is smaller than the alpha value.  We can reject our null hypothesis and assert
#that there's a correlation between the block of vacancies and weapons used. 
```

```python
#Covariance matrix   ---   13. Covariance between NoticeDate and CrimeDate
#Turn columns into ordinal dates and take the covarience
cov_sample = pd.DataFrame(columns=['NoticeDate'])
crime_sample = pd.DataFrame(columns=['CrimeDate'])

cov_bin = []
cri_bin = []

cov_sample["NoticeDate"] = pd.to_datetime(vacant["NoticeDate"])
crime_sample["CrimeDate"] = pd.to_datetime(crime["CrimeDate"])

for i in range(0, len(cov_sample), 1):
    cov_bin.append(cov_sample["NoticeDate"][i].dayofyear)

for i in range(0, len(crime_sample), 1):
    cri_bin.append(crime_sample["CrimeDate"][i].dayofyear)
    
cov_sample = pd.DataFrame(data=cov_bin, columns=['NoticeDate'])
crime_sample = pd.DataFrame(data=cri_bin, columns=['CrimeDate'])
covariant = cov_sample['NoticeDate'].cov(crime_sample['CrimeDate'])

sample = pd.DataFrame(columns=['NoticeDate', 'CrimeDate'])
sample['CrimeDate'] = crime_sample['CrimeDate']
sample['NoticeDate'] = cov_sample['NoticeDate']

sample.corr()
#Explain findings:
#I'm attempting to correlate the dates of crime with the dates of vacancie.  I computed the ordinal dates of each
#date of crime and vacant.  The returned matrix shows that the correlation between NoticeDate and CrimeDate is .0066 
#Which is very close to 0.  Therefore it's safe to say there's no correlation between these two values. 
```
